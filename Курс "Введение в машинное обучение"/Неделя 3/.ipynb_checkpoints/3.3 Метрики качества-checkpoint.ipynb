# Метрики качества классификации - 1

Мы пока не говорили о том, как измерить качество алгоритма. Как понять, хорошо или плохо он работает на той или иной выборке. В этой лекции мы поговорим о мерах качества классификации.

## Постановка задачи

__Дано:__ задача классификации.

$X^l = \{x_1, \ldots, x_l\}$ - выборка;

$y_i = y(x_i) \in \{0, 1\},\ i=1,\ldots, l$ - известные бинарные ответы.

$a: X \rightarrow Y$ - алгоритм, решающая функция, приближающая $y$ на всем множестве объектов $X$.

__Вопрос:__

Как измерить качество $a(x)$ на выборке $X^l$?

## Accuracy

Доля правильных ответов на выборке (accuracy):

$$\frac{1}{l}\sum\limits_{i=1}^{l}[a(x_i) = y_i].$$

- Соответствует интуитивным представлениям о качестве классификации.

- Имеет проблемы с интерпретацией на несбалансированных выборках.

__Пример__ (медицинская диагностика):

- $950$ объектов класса $0$.

- $50$ объектов класса $1$.

- $a(x) = 0$ для всех $x$.

Доля правильных ответов $a(x):\ 95\%$!

Решение: смотреть на базовую долю правильных ответов

$$\mathrm{BaseRate} = \arg \max_{y_0 \in \{0, 1\}}\frac{1}{l}\sum\limits_{i=1}^{l}[y_0 = y_i].$$

В примере: $\mathrm{BaseRate} = 95\%$.

## Матрица ошибок

Ошибки бывают разные:
             
|           | $y=1$             | $y=0$                |
|-----------|-------------------|----------------------|
|$a(x)=1$   | True Positive (TP)|  False Positive (FP) |
|           |                   |                      |
|$a(x)=0$   | False Negative (FN)    |True Negative (TN)    |

Объекты нулевого класса, которые алгоритм отнес к первому относят к "ложным срабатываниям".

$$\mathrm{accuracy} = \frac{TP + TN}{TP + FP + FN + TN}.$$

__Пример:__ задача медицинской диагностики ($y=1$ - больные, $y=0$ - здоровые).

|           |     $1$       | $0$                |
|-----------|-------------------|----------------------|
|$a(x)=1$   |      20           |  50                  |
|           |                   |                      |
|$a(x)=0$   |      5            |1000                  |

Доля правильных ответов: $94.9\%$.

|           |     $1$       | $0$                |
|-----------|-------------------|----------------------|
|$a(x)=1$   |      0           |  0                  |
|           |                   |                      |
|$a(x)=0$   |      25            |1050                  |

Доля правильных ответов константного классификатора: $97.6\%$.

У разных типов ошибки может быть разная _цена_.

Нам не страшно провести несколько дополнительных анализов и понять, что человек здоров. При этом ложный пропуск крайне опасен – мы не определяем больного человека, мы не понимаем, что его нужно лечить. Это крайне нежелательно. Поэтому здесь ложное срабатывание гораздо менее ценно, чем ложный пропуск.

В случаях, когда разные ошибки имеют разную цену, гораздо лучше измерять две другие метрики качества вместо одной: точность и полноту.

## Точность и полнота

Точность (precision) - насколько можно доверять классификатору:

$$\mathrm{precision} = \frac{TP}{TP + FP}.$$

Точность показывает, насколько мы можем доверять классификатору, если он выдает ответ $1$.

|           |     $1$           | $0$                  | $\sum$         | 
|-----------|-------------------|----------------------|----------------|
|$a(x)=1$   |      20           |  50                  |70                |
|           |                   |                      |                | 
|$a(x)=0$   |      5            |1000                  | 1005               |

Точность классификатора: $28.6\%$.

Точность константного классификатора: $0\%$. 

Полнота (recall) - как много объектов класса $1$ находит классификатор:

$$\mathrm{recall} = \frac{TP}{TP + FN}.$$

|           |     $1$       | $0$                |
|-----------|-------------------|----------------------|
|$a(x)=1$   |      20           |  50                  |
|           |                   |                      |
|$a(x)=0$   |      5            |1000                  |
|           |                   |     |
|$\sum$     |   25              |1050 |

Точность классификатора: $80\%$.

Точность константного классификатора: $0\%$.

- Точность и полнота характеризуют разные стороны качества классификатора. 

- Чем выше точность, тем меньше ложных срабатываний. 

- Чем выше полнота, тем меньше ложных пропусков. 

- В зависимости от того, какие ошибки мы меньше хотим допускать, мы можем отдавать предпочтения либо точности, либо полноте. В разных задачах эти предпочтения могут быть разные.

__Пример 1__

Определение мошеннических действий на банковских счетах.

Важнее __полнота__: лучше проверить лишний раз, чем пропустить вредоносные действия.

__Пример 2__

Поиск вражеских самолетов для автоматического уничтожения ракетой.

Важнее __точность__: нельзя допустить стрельбы по своему самолету.

Точность и полнота – очень неплохие метрики. Но при этом их две. В задачах машинного обучения приятнее иметь одну метрику, которую мы будем оптимизировать: минимизировать или максимизировать. Как их объединить?

## Усреднение точности и полноты

Арифметическое среднее:

$$A = \frac{1}{2}(\mathrm{precision} + \mathrm{recall}).$$

- Если $\mathrm{precision} = 0.05,\ \mathrm{recall} = 1$, то $A = 0.525$.

- Если $\mathrm{precision} = 0.525,\ \mathrm{recall} = 0.525$, то $A = 0.525$.

- Первый классификатор — константный, не имеет смысла. 

- Второй классификатор показывает неплохое качество.

```python
from IPython.display import Image
Image(filename='images/3.3 Метрики качества/image_1.png')
```

Минимум:

$$M = \min (\mathrm{precision}, \mathrm{recall}).$$

- Если $\mathrm{precision} = 0.05,\ \mathrm{recall} = 1$, то $M = 0.05$.

- Если $\mathrm{precision} = 0.525,\ \mathrm{recall} = 0.525$, то $M = 0.525$.

- Если $\mathrm{precision} = 0.2,\ \mathrm{recall} = 1$, то $M = 0.2$. 

- Если $\mathrm{precision} = 0.2,\ \mathrm{recall} = 0.3$, то $M = 0.2$.

```python
from IPython.display import Image
Image(filename='images/3.3 Метрики качества/image_2.png')
```

Гармоническое среднее или F-мера:

$$F = \frac{2 \cdot \mathrm{precision} \cdot \mathrm{recall}}{\mathrm{precision} + \mathrm{recall}}.$$

- Если $\mathrm{precision} = 0.05,\ \mathrm{recall} = 1$, то $F = 0.1$.
- Если $\mathrm{precision} = 0.525,\ \mathrm{recall} = 0.525$, то $F = 0.525$.
- Если $\mathrm{precision} = 0.2,\ \mathrm{recall} = 1$, то $F = 0.33$.
- Если $\mathrm{precision} = 0.2,\ \mathrm{recall} = 0.3$, то $F = 0.24$.

```python
from IPython.display import Image
Image(filename='images/3.3 Метрики качества/image_3.png')
```

## Заключение

- Простая мера качества классификации — доля верных ответов.
- Не учитывает цены ошибок.
- Точность и полнота позволяют различать ложные срабатывания и ложные пропуски.
- F-мера — способ усреднения точности и полноты.

# Метрики качества классификации - 2

## Постановка задачи

__Дано:__ задача классификации.

$X^l = \{x_1, \ldots, x_l\}$ - выборка;

$y_i = y(x_i) \in \{0, 1\},\ i=1,\ldots, l$ - известные бинарные ответы.

$b: X \rightarrow \mathbb{R}$ - алгоритм, оценивающий принадлежность $x$ к классу $1$.

__Вопрос:__

Как измерить качество $b(x)$ на выборке $X^l$?

## Откуда берутся оценки принадлежности?

Как правило, классификатор имеет вид

$$a(x) = [b(x) > t].$$

- $b(x)$ — оценка принадлежности к классу $1$.

- $t$ — порог классификации.

Линейный классификатор:

$$a(x) = [⟨w,x⟩ > 0].$$

- Скалярное произведение оценивает расстояние; если объект близко к разделяющей прямой, то мы не уверены, если далеко, то уверены в ответе.

Метод k ближайших соседей:

􏰂􏰃$$a(x) = \Big[\sum\limits_{i=1}^{k}[y^{(i)} =1]>k/2\Big].$$

- Если среди соседей объекта все относятся к одному классу, то он уверен в классификации и выдает высокую оценку; если же среди соседей встречаются оба класса, то оценка понижается.

## Зачем нужны оценки принадлежности?

__Пример:__ кредитный скоринг.
- нужно предсказать, вернет ли клиент кредит;
- сортируем клиентов по оценке вероятности возврата b(x); 
- банк получает ранжированный список;
- порог выбирается в зависимости от стратегии банка; 
- порог может многократно пересматриваться.

__Пример:__ задача определения самолетов противника.
- $b(x)$ оценивает вероятность того, что самолет принадлежит противнику;
- классификатор $a(x) = [b(x) > 0.9]$;
- $\mathrm{precision} = 0.2,\ \mathrm{recall} = 0.7$;
- как понять, в чем проблема — неправильном пороге или плохой функции $b(x)$?

Если бы мы могли оценивать качество алгоритма $b(x)$, то могли бы определить источник проблемы.

Мы разобрались с тем, откуда берутся оценки принадлежности, зачем они нужны. Давайте перейдем к конкретным методам. Первый, о котором мы поговорим, основан на кривой точности и полноты или PR-кривой. 

## PR-кривая

- Отсортируем объекты по возрастанию оценки $b(x)$:

$$b(􏰀x_{(1)})􏰁 \leq 􏰅 \ldots 􏰅 b(􏰀x_{(l)}􏰁).$$

- Переберем все пороги классификации, начав с максимального:
$$t_l =b􏰀(x_{(l)}􏰁),$$
$$\ldots$$
$$t_1 =b(x_{(1)})􏰁,$$
$$t_0 =b(x_{(1)})􏰁−\varepsilon.$$

- Для каждого порога посчитаем точность и полноту.
- Нанесем соответствующую точку в осях "полнота-точность".
- Соединим точки, получив Precision-Recall-кривую.

По скольки точкам строится Precision-Recall-кривая, если все ответы алгоритма $b(x)$ различны?

- (число объектов) + 1. Рассматриваются все возможные способы разбить сортированный список ответов алгоритма на две части. Таких способов ровно (число объектов) + 1, если ответы различны.

```python
from IPython.display import Image
Image(filename='images/3.3 Метрики качества/image_4.png')
```

__Свойства:__

- Левая точка: всегда $(0, 0)$ (все объекты относим к классу $0$);

- Правая точка: $(1,l_+/l)$, $l_+$ — число объектов класса $1$ в выборке;

- Если выборка идеально разделима, то кривая пройдет через точку $(1, 1)$;

- Чем больше площадь под кривой, тем лучше.

__AUC-PRC__ (Area Under Precision-Recall curve) — мера качества для $b(x)$.

## ROC-кривая

ROC — "reciever operating characteristic".

- по оси X: False Positive Rate, доля ошибочных положительных классификаций:

$$FPR = \frac{FP}{FP + TN}.$$


$1 − FPR$ называется _специфичностью_ алгоритма.

- по оси Y: True Positive Rate, доля правильных положительных классификаций:

$$TPR = \frac{TP}{TP + FN}.$$

$TPR$ называется _чувствительностью_ алгоритма

```python
from IPython.display import Image
Image(filename='images/3.3 Метрики качества/image_5.png')
```

Свойства:
- Левая точка: всегда $(0, 0)$ (все объекты относим к классу $0$);

- Правая точка: всегда $(1, 1)$ (все объекты относим к классу $1$);

- Если выборка идеально разделима, то кривая пройдет через точку $(1, 0)$;

- Площадь меняется от $1/2$ до $1$;

- Чем больше площадь под кривой, тем лучше.

__AUC-ROC__ (Area Under ROC-curve) — мера качества для $b(x)$.

## Сравнение

ROC-кривая:

$$FPR = \frac{FP}{FP + TN},\ TPR = \frac{TP}{TP + FN}.$$

- Метрики качества нормируются на размеры классов, ROC-кривая не изменится при перемене соотношения классов.

- Интерпретация: AUC-ROC равен вероятности того, что случайно взятый объект класса $1$ получит оценку выше, чем случайно взятый объект класса $0$. 

- Имеет проблемы при сильном дисбалансе классов.

PR-кривая:

$$precision = \frac{TP}{TP + FP},\ recall = \frac{TP}{TP + FN}.$$

- Точность нормируется на число положительных прогнозов, изменится при перемене соотношения классов.

- Максимально возможная площадь под PR-кривой зависит от соотношения классов.

- Хорошо подходит для измерения качества при сильном дисбалансе классов.

__Пример__

- $100$ объектов класса $1$; 
- $1.000.000$ объектов класса $0$;

- Ранжирование: $50.000$ объектов класса $0$, затем $100$ объектов класса $1$, затем все остальные объекты класса $0$;

Метрики качества: 

- AUC-ROC: $0.95$;
- AUC-PRC: $0.001$. 

Почему так получается?

- Выберем порог, при котором первые $50.095$ объектов относятся к классу $1$;
- $TPR = 0.95$, $FPR = 0.05$; 
- $\mathrm{precision} = 0.0019,\ \mathrm{recall} = 0.95$.

## Заключение

- Работать с оценками принадлежности может быть полезнее, чем с бинарными ответами.

- Две основные метрики качества: AUC-PRC и AUC-ROC. 

- AUC-ROC не зависит от соотношения классов.

# Многоклассовая классификация

## Постановка задачи

__Дано__: $\{x_1,\ldots,x_l\} \subset X$ — выборка;

$y_i = y(x_i) \in \{1,\ldots,K\}$, $i = 1,\ldots,l$ — известные ответы.

__Найти__: $a : X \rightarrow Y$ — алгоритм, решающую функцию, приближающую $y$ на всём множестве объектов $X$.

__Вопросы__:
- Как свести задачу к бинарной классификации?
- Как измерить качество решения?

## One-vs-all

Один против всех.

__Идея__: построить $K$ классификаторов, отделяющих каждый класс от остальных.

Получим $K$ задач бинарной классификации: 
- Объекты: $X^k = X^l$;
- Ответы: $y_i^k = [y_i = k]$;
- Оценка принадлежности: $b_k(x) \in \mathbb{R}$.

Итоговый алгоритм:

$$a(x) = \arg\max_{k =1,\ldots,K} b_k(x).$$ 

## All-vs-all

Каждый против каждого.

__Идея__: построить классификаторы для каждой пары классов.

Получим $K(K − 1)$ задач бинарной классификации:

- Объекты: $X^{km} = \{x \in X^l\ |\ y(x) = k \text{ или } y(x) = m\}$;

- Ответы: $y_{i}^{km} = [y_i = k]$; 

- Оценка принадлежности: $b_{km}(x) \in \mathbb{R}$;

- Симметрия: $b_{km}(x) = −b_{mk}(x)$.

Итоговый алгоритм:

$$a(x) = \arg \max_{k=1,...,K}\sum\limits_{m=1}^{K}b_{km}(x).$$
􏰃  
В выборке три класса. К первому относится $100$ объектов, ко второму $50$ объектов, к третьему — $200$ объектов. На скольки объектах будет обучаться классификатор, разделяющий первый и второй классы, в подходе каждый-против-каждого?

- $150$. Выборка будет состоять из объектов первого и второго классов.

## Сравнение

One-vs-all:

- Линейное число классификаторов, но каждый обучается на полной выборке.

- Может возникнуть проблема с несбалансированными выборками.

All-vs-all:
- Квадратичное число классификаторов, но каждый обучается на небольшой подвыборке.

## Метрики качества

Доля правильных ответов (accuracy):

$$\frac{1}{l}\sum\limits_{i=1}^{l}[a(x_i) = y_i].$$

Матрица ошибок:

|           |     $y=1$           | $\ldots$                  | $y=K$         | 
|-----------|-------------------|----------------------|----------------|
|$a(x)=1$   |      $q_{11}$           |  $\ldots$                  |$q_{1K}$                |
|           |                   |                      |                | 
|$\ldots$   |      $\ldots$            |$\ldots$                  | $\ldots$               |
|||||
|$a(x) = K$|$q_{K1}$|$\ldots$|$q_{KK}$|

где

$$q_{ij} = \sum\limits_{m=1}^{l}[a(x_m) = i][y_m=j].$$

Как обобщить точность, полноту, AUC?

Рассмотрим $K$ задач отделения одного из классов от остальных.

- Микро-усреднение (micro-averaging):
 - Найдем TP, FP, FN, TN для каждой из задач; 
 - Усредним их по всем задачам;
 - Вычислим итоговую метрику.

 - Вклад каждого класса зависит от его размера. 

- Макро-усреднение (macro-averaging):

 - Вычислим итоговую метрику для каждой из задач; 
 - Усредним по всем классам.
 
 - Все классы вносят равный вклад.
 
__Пример__

|           |     $TP$           | $FP$                  | $FN$         | $TN$            |
|-----------|-------------------|----------------------|----------------|------------------|
|$y=1$   |      $900$           |  $120$                  |$100$                |$930$|
|           |                   |                      |                | |
|$y=2$   |      $850$            |$70$                  | $150$               |$980$|
|||||
|$y=3$|$10$|$100$|$40$|$1900$|

Чему равна точность (precision)?

Микро-усреднение:

|     $TP$           | $FP$                  | $FN$         | $TN$            |
|-------------------|----------------------|----------------|------------------|
|      $586.7$           |  $96.7$                  |$96.7$                |$1270$|

Точность: $86\%$.

Макро-усреднение:

|     Класс1           | Класс2                 | Класс3         | 
|-------------------|----------------------|----------------|
|      $88\%$           |  $92\%$                  |$9\%$                |

Точность: $63\%$.

Какой из способов усреднения метрик не зависит от соотношения размеров классов?

- Макро-усреднение. В этом подходе усредняются значения целевой метрики (например, точности). Если эта метрика нормирована (например, практически все изученные нами ранее метрики лежат на отрезке $[0, 1]$), то при усреднении каждый класс будет вносить равный вклад.

## Заключение

- Многоклассовую классификацию можно свести к серии бинарных задач.

- Два подхода: one-vs-all и all-vs-all.

- Вычисление качества также производится через сведение к бинарным задачам.

- Микро-усреднение учитывает наиболее крупные классы.

- Макро-усреднение учитывает все классы одинаково, без учета их размеров.

Подсчитайте число ложных положительных срабатываний относительно первого класса для следующей матрицы ошибок:

|           |     $y=1$           | $y=2$                  | $y=3$         | 
|-----------|-------------------|----------------------|----------------|
|$a(x)=1$   |      $20$           |  $30$                  |$1$                |
|           |                   |                      |                |
|$a(x)=2$   |      $5$            |$50$                  | $70$               |
|||||
|$a(x)=3$|$10$|$0$|$60$|

Ложные положительные срабатывания относительно первого класса — это ситуации, в которых объект относится алгоритмом к классу $1$, но на самом деле относится к другому классу. В данной задаче это число равно $31$.

# Метрики качества классификации

В задачах классификации может быть много особенностей, влияющих на подсчет качества: различные цены ошибок, несбалансированность классов и т.д. Из-за этого существует большое количество метрик качества — каждая из них рассчитана на определенное сочетание свойств задачи и требований к ее решению.

Меры качества классификации можно разбить на две большие группы: предназначенные для алгоритмов, выдающих номера классов, и для алгоритмов, выдающих оценки принадлежности к классам. К первой группе относятся доля правильных ответов, точность, полнота, F-мера. Ко второй — площади под ROC- или PR-кривой.

__Реализация в sklearn__

Различные метрики качества реализованы в пакете sklearn.metrics. Конкретные функции указаны в инструкции по выполнению задания.

### Загрузите файл classification.csv. В нем записаны истинные классы объектов выборки (колонка true) и ответы некоторого классификатора (колонка pred).

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score
from sklearn.metrics import recall_score, f1_score, roc_auc_score, precision_recall_curve
```

```python
df = pd.read_csv('data/classification.csv')
df.head()
```

### Заполните таблицу ошибок классификации. Для этого подсчитайте величины TP, FP, FN и TN согласно их определениям. Например, FP — это количество объектов, имеющих класс 0, но отнесенных алгоритмом к классу 1. Ответ в данном вопросе — четыре числа через пробел.

```python
TP = df[(df["pred"] == 1) & (df["true"] == 1)]
FP = df[(df["pred"] == 1) & (df["true"] == 0)]
FN = df[(df["pred"] == 0) & (df["true"] == 1)]
TN = df[(df["pred"] == 0) & (df["true"] == 0)]

with open('answers/3_3_ans_1.txt', 'w') as f:
    f.write(str(len(TP)) + ' ' + str(len(FP)) + ' ' + str(len(FN)) + ' ' + str(len(TN)))
```

### Посчитайте основные метрики качества классификатора

- Accuracy (доля верно угаданных) — sklearn.metrics.accuracy_score

- Precision (точность) — sklearn.metrics.precision_score

- Recall (полнота) — sklearn.metrics.recall_score

- F-мера — sklearn.metrics.f1_score

```python
acc = accuracy_score(df['true'], df['pred'])
pr = precision_score(df["true"], df["pred"])
rec = recall_score(df["true"], df["pred"])
f1 = f1_score(df["true"], df["pred"])

with open('answers/3_3_ans_2.txt', 'w') as f:
    f.write(str(acc) + ' ' + str(pr) + ' ' + str(rec) + ' ' + str(f1))
```

### Имеется четыре обученных классификатора. В файле scores.csv записаны истинные классы и значения степени принадлежности положительному классу для каждого классификатора на некоторой выборке

- для логистической регрессии — вероятность положительного класса (колонка score_logreg),

- для SVM — отступ от разделяющей поверхности (колонка score_svm),

- для метрического алгоритма — взвешенная сумма классов соседей (колонка score_knn),

- для решающего дерева — доля положительных объектов в листе (колонка score_tree).

Загрузите этот файл.

```python
df2 = pd.read_csv('data/scores.csv')
df2.head()
```

### Посчитайте площадь под ROC-кривой для каждого классификатора. Какой классификатор имеет наибольшее значение метрики AUC-ROC (укажите название столбца)? Воспользуйтесь функцией sklearn.metrics.roc_auc_score.

```python
clf_names = df2.columns[1:]
scores = pd.Series([roc_auc_score(df2["true"], df2[clf]) for clf in clf_names], index=clf_names)

clf = scores.sort_values().index[-1]

with open('answers/3_3_ans_3.txt', 'w') as f:
    f.write(clf)
```

### Какой классификатор достигает наибольшей точности (Precision) при полноте (Recall) не менее 70% ?

Чтобы получить ответ на этот вопрос, найдите все точки precision-recall-кривой с помощью функции sklearn.metrics.precision_recall_curve. Она возвращает три массива: precision, recall, thresholds. В них записаны точность и полнота при определенных порогах, указанных в массиве thresholds. Найдите максимальной значение точности среди тех записей, для которых полнота не меньше, чем 0.7

```python
pr_scores = []
for clf in clf_names:
    pr_curve = precision_recall_curve(df2["true"], df2[clf])
    pr_scores.append(pr_curve[0][pr_curve[1] >= 0.7].max())
    
    
with open('answers/3_3_ans_4.txt', 'w') as f:
    f.write(clf_names[np.argmax(pr_scores)])
```
