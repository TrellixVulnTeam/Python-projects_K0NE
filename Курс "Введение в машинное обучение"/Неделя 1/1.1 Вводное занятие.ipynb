# Формальная постановка задачи машинного обучения

## Задача обучения по прецедентам

- $X$ - множество объектов;

- $Y$ - множество ответов;

- $y : X \rightarrow Y$ - неизвестная зависимость (target function);

__Дано__:

- $\{x_1, \ldots, x_l\} \subset X$ - обучающая выборка (training sample);

- $y_i = y(x_i)$ - известные ответы.

__Найти__:

$a : X \rightarrow Y$ - алгоритм, решающую функцию (decision function), приближающую $y$ на всем множестве $X$.

Весь курс машинного обучения - это конкретизация:

- как задаются объекты и какими могут быть ответы;

- как строить функцию $a$;

- в каком смысле $a$ должен приближать $y$.

## Как задаются объекты
### Признаковое описание

$f_j : X \rightarrow D_j, j = 1, \ldots, n$ - признаки объектов (features).

Типы признаков:

- $D_j = \{0, 1\}$ - _бинарный_ признак $f_j$;

- $|D_j| < \infty$ - _номинальный (категориальный)_ признак $f_j$;

- $|D_j| < \infty$, $D_j$ упорядочено - _порядковый_ признак $f_j$;

- $D_j = \mathbb{R}$ - _количественный_ признак $f_j$.

Пример номинального признака - цвет глаз.

Пример порядкового признака - воинское звание.

Вектор $(f_1(x), \ldots, f_n(x))$ - признаковое описание объекта $x$.

Матрица "объекты-призаки" (feature data)

$$F = \|f_j(x_i)\|_{l\times n} = \begin{pmatrix} f_1(x_1) & \ldots &f_n(x_1) \\ \ldots & \ldots & \ldots \\ f_1(x_l) & \ldots & f_n(x_l) \end{pmatrix}$$

Эта матрица - способ представления обучающей выборки. Строки - объекты, столбцы - признаки. Каждой строчке соответствует ответ.

## Как задаются ответы

### Типы задач

__Задачи классификации__ (classification):

- $Y = \{-1, +1\}$ - _классификация_ на $2$ класса.

- $Y = \{1,\ldots, M\}$ - на $M$ непересекающихся классов (распознавание рукописных символов).

- $Y = \{0, 1\}^M$ - на $M$ классов, которые могут пересекаться (мед. диагностика, у одного пациента сразу много заболеваний).

__Задачи восстановления регрессии__ (regression):

- $Y = \mathbb{R}$ или $Y = \mathbb{R}^m$ (задачи прогнозирования).

__Задачи ранжирования__ (ranking, learning to rank):

- $Y$ - конечное упорядоченное множество.

Задача восстановления регрессии - предсказание цены на жилье по его характеристикам.

## Предсказательная модель

_Модель_ (predictive model) - параметрическое семейство функций $$A = \{a(x) = g(x, \theta)\ |\ \theta \in \Theta\},$$

где $g : X \times \Theta \rightarrow Y$ - фиксированная функция, $\Theta$ - множество допустимых значений параметра $\theta$. Вектор $\theta$ - искомый вектор наших параметров.

__Пример__.

_Линейная модель_ с вектором параметров $\theta = (\theta_1, \ldots, \theta_n),\ \Theta = \mathbb{R}^n$:

- $g(x, \theta) = \sum\limits_{j=1}^{n}\theta_j f_j(x)$ - для регрессии и ранжирования, $Y = \mathbb{R}$;

- $g(x, \theta) = \mathrm{sign}\sum\limits_{j=1}^{n}\theta_j f_j(x)$ - для классификации, $Y = \{-1, +1\}$.

Фактически, линейная модель классификации строит разделяющую гиперплоскость в $n$-мерном пространстве, по одну сторону этой гиперплоскости - объекты одного класса, по другую - другого.

## Этапы обучения и применения модели

__Этап обучения (train)__:

_Метод обучения_ (learning algorithm) $\mu : (X \times Y)^l \rightarrow A$ по выборке $X^l = (x_i, y_i)_{i=1}^{l}$ строит алгоритм $a=\mu(X^l)$: 

$$\begin{bmatrix} \begin{pmatrix} f_1(x_1) & \ldots &f_n(x_1) \\ \ldots & \ldots & \ldots \\ f_1(x_l) & \ldots & f_n(x_l) \end{pmatrix} \xrightarrow{y} \begin{pmatrix} y_1 \\ \ldots \\ y_l \end{pmatrix} \end{bmatrix} \xrightarrow{\mu} a$$

__Этап применения (test)__:

Алгоритм $a$ для новых объектов $x'_1, \ldots, x'_k$ выдает ответы $a(x'_i)$.

$$\begin{pmatrix} f_1(x'_1) & \ldots &f_n(x'_1) \\ \ldots & \ldots & \ldots \\ f_1(x'_k) & \ldots & f_n(x'_k) \end{pmatrix} \xrightarrow{a} \begin{pmatrix} a(x'_1) \\ \ldots \\ a(x'_k) \end{pmatrix}$$

Каким образом выбирать из предсказательной модели тот алгоритм $a(x)$, который будет удовлетворять поставленной задаче? Один из самых распространенных способов - свести задачу обучения к задаче оптимизации. То есть выбрать такой алгоритм, который на большинстве объектов обучающей выборки дает достаточно точные ответы. А как измерять точность ответа алгоритма на одном отдельном объекте? Для этого вводятся функции потерь.

## Функционалы качества

$\mathscr{L}(a,x)$ - _функция потерь_ (loss function) - величина ошибки алгоритма $a \in A$ на объекте $x \in X$.

__Функции потерь для задач классификации__:

- $\mathscr{L}(a,x) = [a(x) \neq y(x)]$ - индикатор ошибки;

__Функции потерь для задач регрессии__:

- $\mathscr{L}(a,x) = |a(x) - y(x)|$ - абсолютное значение ошибки;

- $\mathscr{L}(a,x) = (a(x) - y(x))^2$ - квадратичная ошибка.

_Эмпирический риск_ - функционал качества алгоритма $a$ на $X^l$:

$$Q(a, X^l) = \frac{1}{l}\sum\limits_{i=1}^{l}\mathscr{L}(a,x_i).$$

Мы ищем такое $\theta$, которое будет давать минимум эмпирического риска.

## Сведение задачи обучения к задаче оптимизации

_Минимизация эмпирического риска_ (empirical risk minimization):

$$\mu(X^l) = \arg\min_{a\in A}Q(a, X^l).$$

__Пример__: _метод наименьших квадратов_ $(Y = \mathbb{R}, \mathscr{L} \text{ квадратична})$:

$$\mu(X^l) = \arg\min_{\theta}\sum\limits_{i=1}^{l}(g(x_i, \theta) - y_i)^2.$$

## Понятие обобщающей способности (generalization performance)

- найдем ли мы закон природы или _"переобучимся",_ то есть подгоним функцию $g(x_i, \theta)$ под заданные точки?

- будет ли $a=\mu(X^l)$ приближать функцию $y$ на всем $X$?

- будет ли $Q(a, X^k)$ мало на новых данных - _контрольной выборке_ $X^k = (x'_i, y'_i)_{i=1}^{k},\ y'_i = y(x_i)$?

## Пример. Переобучение полиномиальной регрессии

Зависимость $y(x) = \frac{1}{1 + 25x^2}$ на отрезке $x \in [-2, 2]$.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-2, 2.01, 0.01)
y = [1/(1 + 25 * el ** 2) for el in x]

plt.plot(x, y)
```

Признаковое описание $x \rightarrow(1, x^1, x^2, \ldots, x^n)$. 

Модель полиномиальной регрессии $$a(x, \theta) = \theta_0 + \theta_1 x + \ldots + \theta_nx^n - \text{полином степени $n$}.$$

Обучение методом наименьших квадратов: $$Q(a, X^l) = \sum\limits_{i=1}^{l}(\theta_0 + \theta_1 x_i + \ldots + \theta_nx_i^n - y_i)^2 \rightarrow \min_{\theta_0,\ldots,\theta_n}.$$

Обучающая выборка: $X^l = \{x_i = 4\frac{i-1}{l-1} - 2\ |\ i=1,\ldots,l\}.$

Контрольная выборка: $X^k = \{x_i = 4\frac{i-0.5}{l-1} - 2\ |\ i=1,\ldots,l-1\}.$

Такая модель может быть неустойчивой по отношению к шумам, к погрешностям измерений, и она уже не восстанавливает закон природы, значит, нужно как-то ограничивать сложность модели.

__Переобучение__ - это ситуация, в которой алгоритм показывает хорошее качество на обучающей выборке, но при этом плохо работает на новых данных.

Для того чтобы измерять такие эффекты, принято использовать функционалы, которые непосредственно измеряют качество построенной модели на тех данных, на которых эта модель не обучалась.

## Эмпирические оценки обобщающей способности

Самый простой способ — поделить выборку между обучением и контролем. $X^l$ - это обучение, $X^k$ - это контроль. Исходно у нас было $(l + k)$ точек, и мы их как-то разбили на $2$ части. На одной обучаемся, на второй проверяемся. 

- Эмпирический риск на тестовых данных (hold-out):

$$HO(\mu, X^l, X^k) = Q(\mu(X^l), X^k) \rightarrow \min.$$

Но этот функционал (эмпирический риск на тестовых данных) существенно зависит от разбиения. Можно так неудачно выбрать разбиение, что оценка будет субъективной, смещенной.

Чтобы этого не происходило, делают много разбиений. Простейший способ это устроить — каждый раз выделять по одному контрольному объекту, а на всей оставшейся выборке проводить обучение.

- Скользящий контроль (leave-one-out), $L = l + 1$:

$$LOO(\mu, X^L) = \frac{1}{L}\sum\limits_{i=1}^{l}\mathscr{L}(\mu(X^L \setminus \{x_i\}), x_i) \rightarrow \min.$$

Минус этого способа в том, что обучение придется проводить столько раз, сколько было объектов обучающей выборки, это может быть просто долго.

Третий способ — это всевозможные методы кросс-проверки, или кросс-валидации, когда мы каким-то образом устраиваем много $N$ разбиений выборки на обучение и контроль и каждый раз на одной части обучаемся, на другой проверяем результат и усредняем по всевозможным разбиениям.

- Кросс-проверка (cross-validation) по $N$ разбиениям, $X^L = X_n^l \sqcup $X_n^k, L = l+k$:

$$CV(\mu, X^L) = \frac{1}{N}\sum\limits_{n=1}^{N} Q(\mu(X_n^l), X_n^k) \rightarrow \min.$$


Этот способ уже менее зависим от того, какие именно разбиения будут выбраны. Можно выбирать случайные разбиения.
