# Объекты и ответы

- $x$ - объект (то, для чего надо сделать предсказание)
- $\mathbb{X}$ - пространство объектов
- $y = y(x)$ - ответ
- $\mathbb{Y}$ - пространство ответов

# Признаки
- признак - числовая характеристика, понятная компьютеру
- $x = (x^1, x^2, \ldots, x^d)$ - признаковое описание 
- $x^i$ - число, элемент множества

# Обучающая выборка
- Машинное обучение - извлечение закономерностей из примеров
- $\textbf{X} = (x_i, y_i)_{i=1}^l$, $x_i$ - $i$-ый объект, $y_i$ - $i$-ый истинный ответ на нем

# Алгоритм
- $a(x)$ - алгоритм, модель
- функция из $\mathbb{X}$ в $\mathbb{Y}$

- Пример: линейный алгоритм 
- $a(x) = \mathrm{sign}(w_0 + w_1x^1 + \ldots + w_dx^d)$

# Функционал ошибки - характеристика нашего алгоритма
- $a(x) = 1$ - не подходит
- $Q(a, \textbf{X})$ - ошибка алгоритма $a$ на выборке $\textbf{X}$
- Пример: доля неправильных ответов

# Обучение

Задача обучения - поиск такого $a$, на котором достигается минимум функционала ошибки
- $Q(a, \textbf{X}) \rightarrow \min$

# Где ищем такое $a$?
- $\mathbb{A}$ - семейство алгоритмов
- Решающие пни - $\mathbb{A} = \{[x^j < t]\ |\ \forall j,t\}$ 
- [True] == 1, [False] == 0

# Машинное обучение
- Каким выбрать функционал ошибки?
- Какое взять семейство алгоритмов?
- Как обучить алгоритм?
- Как сформировать признаки?
- Как подготовить данные?

# Виды обучения
- Обучение с учителем - есть и объекты и ответы
- Обучение без учителя - есть только объекты
- Частичное обучение - объекты, некоторые с ответами
- Активное обучение - на каком объекте получить ответ?

# Обучение на размеченных данных

- $\textbf{X} = (x_i, y_i)_{i=1}^l$ - обучающая выборка
- $Q(a, \textbf{X}) \rightarrow \min$ - обучение с учителем
- Какие бывают ответы?

- Тип задачи определяется пространством ответов, т. е. $\mathbb{Y}$

## Задача классификации
- $\mathbb{Y} = \{0, 1\}$ - бинарная классификация
- $\mathbb{Y} = \{0, 1, \ldots, \ K\}$ - многоклассовая классификация

## Регрессия
- $\mathbb{Y} = \mathbb{R}$
- Предсказание роста по весу

## Ранжирование

# Обучение без учителя

## Кластеризация
- Найти группы похожих объектов
- Не знаем количество кластеров
- Не знаем правильных ответов

## Визуализация
- Нарисовать многомерную выборку
- Должна отражать структуру объектов
- Должно быть красиво

## Поиск аномалий
- Определить объект, который не похож на большинство
- Примеров аномалий - мало или нет вообще

# Признаки в машинном обучении

## Признаки
- Описывают объекты
- $D_j$ - множество значений $j$-ого признака

## Бинарные признаки
- $D_j = \{0, 1\}$ 

## Вещественные признаки
- $D_j = \mathbb{R}$ 

## Категориальные признаки
- $D_j$ - неупорядоченное множество
- Цвет глаз

## Порядковые признаки
- $D_j$ - упорядоченное множество
- Роль в фильме

## Множествозначные признаки
- (set-valued)
- $D_j$ - множество всех подмножеств некоторого множества
- Какие фильмы посмотрел пользователь?

## Выбросы
- Значения признаков, не похожие на большинство
