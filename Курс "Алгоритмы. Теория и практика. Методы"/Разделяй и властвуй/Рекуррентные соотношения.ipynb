# Основная теорема о рекуррентных соотношениях

Рассмотрим алгоритм, основанный на методе "разделяй и властвуй", который для решения задачи размера $n$ делает $a$ рекурсивных вызовов для задач размера $n/b$ и тратит время $O(n^d)$ на то, чтобы подготовить рекурсивные вызовы и чтобы собрать из их ответов ответ для исходной задачи: $$T(n) = aT(\Big\lceil \frac{n}{b} \Big\rceil) + O(n^d),$$ где $a > 0, b > 1, d \geq 0$. Тогда $$T(n) \begin{cases}
O(n^d)&\text{если $d>\log_b a$}\\
O(n^d\log n)&\text{если $d=\log_b a$}\\
O(n^{\log_b a})&\text{если $d<\log_b a$}\\
\end{cases}$$

## Доказательство

Можем считать, что $n$ - это какая-то степень $b$ $(n = b^k)$.
