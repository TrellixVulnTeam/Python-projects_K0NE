```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students_performance = pd.read_csv('StudentsPerformance.csv')
students_performance.head()
```

```python
students_performance = students_performance.rename(
                        columns=
                            {'parental level of education':'parental_level_of_education',
                            'test preparation course' : 'test_preparation_course',
                            'math score' : 'math_score',
                            'reading score' : 'reading_score',
                            'writing score' : 'writing_score'})
students_performance.head()
```

```python
students_performance.math_score.hist()
```

```python
students_performance.plot.scatter(x='math_score', y='reading_score')
```

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance)
```

### Если регрессионные прямые не нужны - fit_reg=False

### Задача
Постройте гистограмму, отражающую скольким героям сколько ролей приписывается (по мнению Valve, конечно) и напишите какое число ролей у большинства героев.

```python
df_3 = pd.read_csv('dota_hero_stats.csv')
df_3.head()
```

```python
lengths = [len(r.split(',')) for r in df_3.roles]
df_3['lengths'] = lengths
df_3.head()
```

```python
sns.distplot(df_3.lengths)
print(df_3.lengths.mode())
```
